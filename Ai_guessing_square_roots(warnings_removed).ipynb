```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import joblib
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
db100 = pd.read_csv("square_roots_100.csv")
db1000 = pd.read_csv("square_roots_1000.csv")
db10000 = pd.read_csv("square_roots_10000.csv")
db100000 = pd.read_csv("square_roots_100000.csv")
```

```python
print(db100.head())
print(db1000.head())
print(db10000.head())
print(db100000.head())

print(db100.describe())
print(db1000.describe())
print(db10000.describe())
print(db100000.describe())
```

```python
feature100 = db100[["Number"]]
target100 = db100[["Square Root"]]

feature1000 = db1000[["Number"]]
target1000 = db1000[["Square Root"]]

feature10000 = db10000[["Number"]]
target10000 = db10000[["Square Root"]]

feature100000 = db100000[["Number"]]
target100000 = db100000[["Square Root"]]
```

```python
x_train100, x_test100, y_train100, y_test100 = train_test_split(feature100, target100, test_size=0.2, random_state=42)

x_train1000, x_test1000, y_train1000, y_test1000 = train_test_split(feature1000, target1000, test_size=0.2, random_state=42)

x_train10000, x_test10000, y_train10000, y_test10000 = train_test_split(feature10000, target10000, test_size=0.2, random_state=42)

x_train100000, x_test100000, y_train100000, y_test100000 = train_test_split(feature100000, target100000, test_size=0.2, random_state=42)
```

```python
model100 = RandomForestRegressor()

model1000 = RandomForestRegressor()

model10000 =RandomForestRegressor()

model100000 = RandomForestRegressor()
```

```python
model100.fit(x_train100, y_train100)

model1000.fit(x_train1000, y_train1000)

model10000.fit(x_train10000, y_train10000)

model100000.fit(x_train100000, y_train100000)
```

```python
print("Model 100p accuracy:", model100.score(x_train100, y_train100))
print("Model 1000p accuracy:", model1000.score(x_train1000, y_train1000))
print("Model 10000p accuracy:", model10000.score(x_train10000, y_train10000))
print("Model 100000p accuracy:", model100000.score(x_train100000, y_train100000))
```

```python
prediction100 = model100.predict(x_test100)
mse100 = mean_squared_error(y_test100, prediction100)
r2100 = r2_score(y_test100, prediction100)

prediction1000 = model1000.predict(x_test1000)
mse1000 = mean_squared_error(y_test1000, prediction1000)
r21000 = r2_score(y_test1000, prediction1000)

prediction10000 = model10000.predict(x_test10000)
mse10000 = mean_squared_error(y_test10000, prediction10000)
r210000 = r2_score(y_test10000, prediction10000)

prediction100000 = model100000.predict(x_test100000)
mse100000 = mean_squared_error(y_test100000, prediction100000)
r2100000 = r2_score(y_test100000, prediction100000)

print(f"MSE score for 100p {mse100}")
print(f"R2 score for 100p {r2100}")

print(f"MSE score for 1000p {mse1000}")
print(f"R2 score for 1000p {r21000}")

print(f"MSE score for 10000p {mse10000}")
print(f"R2 score for 10000p {r210000}")

print(f"MSE score for 100000p {mse100000}")
print(f"R2 score for 100000p {r2100000}")
```

```python
while True:
  user_input = input("Please select a number that you would like these models to guess the square root of (q for quit) >> ")
  user_input = user_input.lower().strip()

  if user_input == 'q':
    print("exiting now...")
    break

  try:
    number = int(user_input)
    p100 = model100.predict([[number]])[0]
    p1000 = model1000.predict([[number]])[0]
    p10000 = model10000.predict([[number]])[0]
    p100000 = model100000.predict([[number]])[0]

    actual = number ** 0.5

    error100 = abs(p100 - actual)
    error1000 = abs(p1000 - actual)
    error10000 = abs(p10000 - actual)
    error100000 = abs(p100000 - actual)

    print(f"Model 100p guessed {p100}")
    print(f"Model 100p error was {error100}")
    print(f"Model 1000p guessed {p1000}")
    print(f"Model 1000p error was {error1000}")
    print(f"Model 10000p guessed {p10000}")
    print(f"Model 10000p error was {error10000}")
    print(f"Model 100000p guessed {p100000}")
    print(f"Model 100000p error was {error100000}")
  except ValueError:
    print("⚠️ Please enter a valid number.\n")
```
