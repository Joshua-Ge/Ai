```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import re
import warnings
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv("/content/South Australia housing prices.csv")
data = data.dropna()
print(data.describe())
print(data.head())
```

```python
data["City_encoded"], city_encoder = pd.factorize(data["City"])
data["Suburb_encoded"], suburb_encoder = pd.factorize(data["Suburb"])

for i in data["Median 1Q 2023"]:
  data["clean_median2023"] = i.replace(",",'')
```

```python
feature = data[["City_encoded","Suburb_encoded"]]
target = data["clean_median2023"]
```

```python
warnings.filterwarnings("ignore")
x_train, x_test, y_train, y_test = train_test_split(feature,target,test_size=0.2, random_state=42)
```

```python
modelLinearRegression = LinearRegression()
modelRandomForestRegrssion = RandomForestRegressor()
```

```python
modelLinearRegression.fit(x_train,y_train)
modelRandomForestRegrssion.fit(x_train,y_train)
```

```python
results = modelLinearRegression.score(x_train,y_train)
print(f"The modelLinearRegression score is: {results}")

results2 = modelRandomForestRegrssion.score(x_train,y_train)
print(f"The modelRandomForestRegrssion score is: {results2}")
```

```python
def predict_price(city_name, suburb_name, model, city_encoder, suburb_encoder):
    """
    Predicts the housing price for a given city and suburb using the trained model.
    """
    try:
        city_encoded = city_encoder.get_loc(city_name)
        suburb_encoded = suburb_encoder.get_loc(suburb_name)
    except KeyError:
        print("City or Suburb not recognised. Please check spelling.")
        return None

    input_df = pd.DataFrame([[city_encoded, suburb_encoded]], columns=["City_encoded", "Suburb_encoded"])
    prediction = model.predict(input_df)
    return prediction[0]

# Example query
predicted_price = predict_price("ADELAIDE HILLS", "ADELAIDE", modelRandomForestRegrssion, city_encoder, suburb_encoder)
print(f"Predicted housing price for ADELAIDE, ADELAIDE HILLS: ${predicted_price:,.2f}")
```
