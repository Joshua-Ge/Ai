```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

```python
db =  pd.read_csv("/content/newtons_second_law.csv")
```

```python
print(db.describe())
print(db.head())
```

```python
feature = db[["mass_kg","acceleration_m_s2"]]
target = db["force_N"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=24)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
results = model.score(x_test,y_test)
results = results * 100
results = round(results, 2)
print(f"The results of the linear regression model is {results}%")

predict = model.predict(x_test)
MSE = mean_squared_error(y_test,predict)
print(f"The MSE results are {MSE}")
```

```python
while True:
  user_input_mass = input("What is the mass ").lower().strip()

  if user_input_mass == "q":
    print("exiting now...")
    break

  user_input_accel = input("what is the acceleration ").lower().strip()

  try:

    mass = float(user_input_mass)
    accel = float(user_input_accel)

    newtons = model.predict([[mass,accel]])[0]
    real = mass * accel
    print(f"Result: {newtons}")
    print(f"Error: {abs(real - newtons)}")

  except ValueError:
    print("⚠️ Please enter a valid number.\n")
```
